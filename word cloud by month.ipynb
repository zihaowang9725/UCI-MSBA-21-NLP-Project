# Group content by quater

```python
#convert to date type
df['pub_date_after'] = pd.to_datetime(df['pub_date_after'])
```

## Q1

```python
#greater than the start date and smaller than the end date
q1 = (df['pub_date_after'] >= '2020-02-01') & (df['pub_date_after'] < '2020-03-01')

#Select the sub-DataFrame
q1_df = df.loc[q1]

#convert it to a list 
q1_df_list = q1_df["headline"].values.tolist()

#convert list values to a single stringf
q1_df_list_final = ' '.join([str(elem) for elem in q1_df_list]) #list comprehension
```

## Q2

```python
#greater than the start date and smaller than the end date
q2 = (df['pub_date_after'] > '2020-03-31') & (df['pub_date_after'] <= '2020-06-30')

#Select the sub-DataFrame
q2_df = df.loc[q2]

#convert it to a list 
q2_df_list = q2_df["headline"].values.tolist()

#convert list values to a single stringf
q2_df_list_final = ' '.join([str(elem) for elem in q2_df_list]) #list comprehension
```
